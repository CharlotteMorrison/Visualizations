```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.style.use('ggplot')
data = np.random.randn(50)
```

```python
styles = plt.style.available
print(plt.style.available)
```

```python
len(styles)
```

```python
index = 1

fig = plt.figure(figsize=(14, 28))

for style in styles:
    plt.style.use(style)
    fig.add_subplot(9, 3, index)
    index += 1
    plt.plot(data)
```

## Temporary Styling

Change the style for a block of code, but leave the global formatting intact.

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

### Dynamic rc settings
Settings can be dynamically changed in a script. (from default)

```python
mpl.rcParams['lines.linewidth']=2
mpl.rcParams['lines.color']='r'
plt.plot(data)
```

matplotlib.rc can be used to modify multiple settings at once

```python
mpl.rc('lines', linewidth=4, color='g')
plt.plot(data)
```
