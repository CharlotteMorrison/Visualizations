```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
%matplotlib inline
```

# Importing Image data
Matplotlib only supports .png, it falls back on the Pillow library.

```python
# load the image
img = mpimg.imread('stinkbug.png')
print(img)
```

```python
imgplot = plt.imshow(img)
```

### Applying Pseudocolor
To enhance contrast, only applies to one channel.

```python
lum_img = img[:, :, 0]
plt.imshow(lum_img)
```

```python
# green is virdid, can choose other color schemes
plt.imshow(lum_img, cmap='hot')
```

```python
# colormap can be changed on existing  plot objects
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral')
```

### Add a color bar for reference

```python
imgplot = plt.imshow(lum_img)
plt.colorbar()
```

### Create a histogram of colors

```python
plt.hist(lum_img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
plt.show()
```

### The image colors can be clipped
This removes areas of little interest, or could remove sections based on intensity (everything darker is considered black...)

```python
imgplot = plt.imshow(lum_img, clim=(0.0, 0.7))
```

The clim command can also be specified using a returned object.  
see below:

```python
fig = plt.figure()
a = fig.add_subplot(1, 2, 1)

imgplot = plt.imshow(lum_img)
a.set_title('Before')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')

a = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(lum_img)
imgplot.set_clim(0.3, 0.7)
a.set_title('After')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')
```

##  Array Interpolation Schemes
Figures out what the color of a pixel should be.  

### Resizing Images
THe default is bilinear.

```python
from PIL import Image

img = Image.open('stinkbug.png')
img.thumbnail((64, 64), Image.ANTIALIAS) # resize in place
imgplot = plt.imshow(img)
```

#### different methods of smoothing/interpolation

```python
fig = plt.figure()

a = fig.add_subplot(1, 3, 1)
imgplot = plt.imshow(img)
a.set_title('bilinear')

a = fig.add_subplot(1, 3, 2)
imgplot = plt.imshow(img, interpolation="nearest")
a.set_title('nearest')

a = fig.add_subplot(1, 3, 3)
imgplot = plt.imshow(img, interpolation="bicubic")
a.set_title('bicubic')
```
